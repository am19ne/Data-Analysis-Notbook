```python
import pandas as pd

data = pd.read_csv('C:/Users/ADMIN/Documents/iris_data.csv') 
data.columns = ['sepal longueur', 'sepal largeur', 'petal longueur', 'petal largeur', 'classe']

data.head()
```

```python
data.info()
```

```python
data.describe(include="all")
```

```python
col = 'sepal longueur'
min = data[col].min()
max = data[col].max()
mediane = data[col].median()
mode = data[col].mode()[0]
midrange = (max - min)/2
ecart_type = data[col].std()
print('col:',col,
    '\n\tmin:', min,
    'max:',max,
    'mediane:', mediane,
    'mode:', mode,
    'midrange:', midrange,
    'ecart-type:', ecart_type)
```

```python
mode = data[col].mode()
mode
```

```python
from pandas.api.types import is_numeric_dtype

for col in data.columns:
    if is_numeric_dtype(data[col]):
        print('%s:' % (col))
        print('\t Moyenne = %.2f' % data[col].mean())
        print('\t Ecart-type = %.2f' % data[col].std())
        print('\t Minimum = %.2f' % data[col].min())
        print('\t Maximum = %.2f' % data[col].max())
```
